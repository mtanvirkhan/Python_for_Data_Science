## Reading data from multiple csv files and prepare dataframe according to those csv files

```python
# importing pandas
import pandas as pd
```

```python
# dataframe is prepared from single csv file
pd.read_csv('./league_data/2425EC.csv')
```

```python
# prepare the seasons data
seasons = []
for season in range(15, 25):
    season = str(season) + str(season + 1)
    print(season)
    seasons.append(season)
    print(seasons)
```

```python
# source file links are created and total count
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
sum = 0
for season in seasons:
    for league in leagues:
        print(f'file {sum}'+': '+'./league_data/'+''+season+league+'.csv')
        sum = sum + 1
print(f'Total files count: {sum}')
```

```python
# dataframes creation from multiple csv files according to season and leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
dataframes = []
for season in seasons:
    for league in leagues:
        dataframes.append(pd.read_csv('./league_data/'+''+season+league+'.csv', encoding='latin1'))
print(f'Total dataframes data: {len(dataframes)}')
```

```python
# display all the dataframes data one by one season-league-dataframe format
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
# Create season-league pairs in same order as files were read
season_league_pairs = []
for season in seasons:
    for league in leagues:
        season_league_pairs.append((season, league))
        # print(season, league)
print(f'dataframe length: {len(dataframes)}')
print(f'season_league_pairs length: {len(season_league_pairs)}')

# Loop over dataframes and season-league pairs
for i, (dataframe, (season, league)) in enumerate(zip(dataframes, season_league_pairs)):
    print(f'\nðŸ“„ DataFrame {i}: Season {season}, League {league}\n')
    print(dataframe.head()) # first 5 wors data
```
