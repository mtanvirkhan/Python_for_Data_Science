```python
# Display the dataframe after adding the column
df_exams
```

```python
# import numpy
import numpy as np
```

```python
# gearating array data range from 0 to 1000 and check the length
array_data = np.arange(0, 1000)
print(array_data)
print(len(array_data))
```

```python
# assign this faulty array data in the frame with language scre column
df_exams['language score'] = array_data
# showing data
df_exams
```

```python
# creating random integer (0 to 100) for array data range of 1000
random_data = np.random.randint(1, 100, 1000) # for random integer number
len(random_data) # check the length of the array
```

```python
# assign this random data to df_exams language score column
df_exams['language score'] = random_data
# show the dataframe
df_exams
```

```python
# min value inclusive and high value exclusive
print(min(random_data))
print(max(random_data))
```

```python
# creating random floating value from the range of 1 to 100 for 1000 records
rand_float_data = np.random.uniform(1, 100, 1000)
print(len(rand_float_data))
rand_float_data
```

```python
# assign to language score column with this data
df_exams['language score'] = rand_float_data
# display the dataframe in 2 points decimal value
df_exams.round(2)
```
