# The set_index() and sort_index() methods

```python
import pandas as pd
```

```python
# read the student profile file and prepare the data frame
df_exams = pd.read_csv('./StudentsPerformance.csv')
#display the dataframe
df_exams
```

```python
import numpy as np
import random
```

```python
# random 1000 integer data are prepared
random_data = np.random.randint(0, 1000, 1000)
random_data
```

```python
# creating non repetitive values for the index
new_index_shuffle = np.arange(0, 1000)
# display the created data which is sequentially created
new_index_shuffle
```

```python
# creating the type of these 2 variable
print(type(random_data))
print(type(new_index_shuffle))
```

## Create index

```python
# shuffling the indexes
random.shuffle(new_index_shuffle)
new_index_shuffle
```

```python
# new column is created and shuffle data is inserted
df_exams['new_index_shuffle'] = new_index_shuffle
# show the dataframe with new column
df_exams
```

```python
# new column is created and random data is inserted
df_exams['random_data'] = random_data
# show the dataframe with new column
df_exams
```

## Set index

```python
# set new_index column as index
df_exams.set_index(['new_index_shuffle'], inplace=True)
# inplace=True: This parameter determines whether to return a new DataFrame or to modify the existing DataFrame. When inplace=True, 
# the operation modifies the DataFrame df_exams directly. This means that the current DataFrame will now have its index set to the values 
# from the new_index column, and no new DataFrame is created. If inplace were set to False or omitted, the method would return a new DataFrame 
# with the changes, leaving the original DataFrame unaltered.
```

```python
# check the df_exams dataframe
# pd.set_option('display.max_rows', None)
df_exams
```

# Sort index

```python
# sort dataframe by new index
df_exams.sort_index()
```

```python
# sort dataframe by new index in descending order
df_exams.sort_index(ascending=False, inplace=True)
df_exams
```
