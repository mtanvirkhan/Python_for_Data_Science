# Creating a dataframe from an array
## Option 1 using numpy library

```python
# Creating an array
data = np.array([['a', 'b', 'c'], [1, 4, 4], [2, 5, 0], [3, 6, 9]])
print(data)
```

```python
# Creating a dataframe
df = pd.DataFrame(data)
df
```

```python
# Creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3', 'row4'], columns=['col1', 'col2', 'col3'])
df
```

```python
# Showing the dataframe
print(df)
```

# Creating dataframe from dictionary

## example1

```python
# Dictionary data
dic_data = {
    'name': ['a', 'b', 'c', 'd', 'e'],
    'income': [10000, 20000, 35000, 44500, 0],
    'age': [25, 35, 40, 55, 0]
}
# Showing dictionary data
dic_data
```

```python
# Creating dataframe
df = pd.DataFrame(dic_data)
# Showing dataframe
df
```

```python
# Creating dataframe
df = pd.DataFrame(dic_data, index=['data1', 'data2', 'data3', 'data4', 'data5'])
# Showing dataframe
df
```

```python
# Creating dataframe with column name added which lead to invalid information
df = pd.DataFrame(dic_data, index=['data1', 'data2', 'data3', 'data4', 'data5', 'data6'], columns=['Name', 'Income', 'Age'])
# Showing dataframe
df
```

## example 2

```python
# List preparation
states = ['California', 'Texas', 'Florida', 'New York']
population = [39613493, 29730311, 21944577, 19299981]
# Show list data
print(states)
print(population)
```

```python
# Storing lists data within a dictionary
dict_states = {
    'States': states,
    'Population': population
}
# Show the dictionary data
dict_states
```

```python
# Creating the dataframe
df_population = pd.DataFrame(dict_states)
# Showing the dataframe
df_population
```
