```python
import pandas as pd

seasons = ['s1', 's2', 's3', 's4', 's5']
episodes = ['e1', 'e2', 'e3', 'e4', 'e5', 'e6', 'e7', 'e8', 'e9', 'e10', 'e11', 'e12']
perticipants = ['Tanvir', 'Wafia', 'Waziha', 'Adiyat', 'Shormi', 'Amma']
perticipants_occupation = ['Service', 'Student', 'Student', 'Student', 'Housewife', 'Housewife']
dict = {
    'perticipants': perticipants,
    'perticipants_occupation': perticipants_occupation
}

# preapare a data set of 60 dataframes with same data
dfs = []
for i in range(0, 60):
    dfs.append(pd.DataFrame(dict))

# season-episode pair data
season_episode = []
for season in seasons:
    for episode in episodes:
       season_episode.append((season, episode))
       #print(season, episode)

# print the lengths
print(f'length of season_episode: {len(season_episode)}')
print(f'Length dfs: {len(dfs)}')

for i, (dfs, (season, episode)) in enumerate(zip(dfs, season_episode)):
    print(f'dataframe{i}: ', f'season {season}', f'episode {episode}', '\n' , dfs)
```
