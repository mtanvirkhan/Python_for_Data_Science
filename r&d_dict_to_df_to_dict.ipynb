```python
import pandas as pd
```

```python
name = ['abc', 'def', 'ghi', 'jkl', 'mno', 'pqr']
age = [23, 45, 34, 25, 30, 29]
department = ['HR', 'Finance', 'IT', 'Marketing', 'Sales', 'Operations']
tax = [0.1, 0.2, 0.3, 0.15, 0.25, 0.05]
address = ['addr1', 'addr2', 'addr3', 'addr4', 'addr5', 'addr6']
book1 = ['book1.1', 'book1.2', 'book1.3', 'book1.4', 'book1.5', 'book1.6']
book2 = ['book2.1', 'book2.2', 'book2.3', 'book2.4', 'book2.5', 'book2.6']
book3 = ['book3.1', 'book3.2', 'book3.3', 'book3.4', 'book3.5', 'book3.6']
book4 = ['book4.1', 'book4.2', 'book4.3', 'book4.4', 'book4.5', 'book4.6']
book5 = ['book5.1', 'book5.2', 'book5.3', 'book5.4', 'book5.5', 'book5.6']

dict_to_df = {
    'Name': name,
    'Age': age,
    'Department': department,
    'Tax': tax,
    'Address': address,
    'Salary': [1000, 2000, 3000, 4000, 5000, 6000],
    'Experience': [1, 2, 3, 4, 5, 6],
    'Book1': book1,
    'Book2': book2,
    'Book3': book3,
    'Book4': book4,
    'Book5': book5
}

df = pd.DataFrame(dict_to_df)

```
