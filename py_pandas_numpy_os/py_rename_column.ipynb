```python
# import pandas pd
import pandas as pd
```

```python
# read the csv file and build up the dataframe
df_exams = pd.read_csv('./StudentsPerformance.csv')
# show the data
df_exams
```

# Rename column

```python
# rename column and overwrite values of the dataframe
# the rename format structure is like dictionary
# single/ multiple names can be changed accordingly
# this change will not have any impact on main df_exams dataframe
df_exams_rename_column = df_exams.rename(
    columns = {
    'gender': 'Gender',
    'race/ethnicity': 'Race/Ethnicity',
    'parental level of education': 'Parental level of education',
    'lunch': 'Lunch',
    'test preparation course': 'Test preparation course'
})
df_exams_rename_column
```

```python
# original dataframe
df_exams
```

# Rename column using inplace method

```python
# rename column and overwrite values of the dataframe
# the rename format structure is like dictionary
# single/ multiple names can be changed accordingly
# this change will have any impact on main df_exams dataframe for using inframe method
df_exams_rename_column_inplace = df_exams.rename(
    columns = {
    'gender': 'Gender',
    'race/ethnicity': 'Race/Ethnicity',
    'parental level of education': 'Parental level of education',
    'lunch': 'Lunch',
    'test preparation course': 'Test preparation course'
}, inplace = True)
df_exams_rename_column_inplace
```
