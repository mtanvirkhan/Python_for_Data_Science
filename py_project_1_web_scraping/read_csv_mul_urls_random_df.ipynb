## Reading csv files from multiple random URLs 

```python
import pandas as pd
```

```python
# Prepare the csv links with under e list
http = 'https://'
csv_file_links = [
    'archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
    'raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv',
    'web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv',
    'archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',
    'archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',
    'archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'   
]
frames = []
for i, link in enumerate(csv_file_links):
    df = pd.read_csv(http + link, encoding='unicode_escape')
    df.insert(0, 'inserted data data', 'Collected information from website')
    # Append the result in a list
    frames.append(df)
```

```python
# check the length of the preared array
print(len(frames))
# display all frames data index by index
for i in range(0, len(frames)):
    print(f'frame[{i}]: {frames[i]}' + '\n')
```
