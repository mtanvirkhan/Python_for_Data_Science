## Operations on dataframe

```python
import pandas as pd
```

```python
# reading csv file and build dataframe
df_exams = pd.read_csv('../StudentsPerformance.csv')
# show dataframe
round(df_exams, 3)
```

```python
# check the statistical information of dataframe by using describe method
df_exams.describe().round(5)
```

## Math operatons in column

### sum, count, mean, std, max and min operation

```python
# sum
df_exams['math score'].sum()
```

```python
# count
df_exams['math score'].count()
```

```python
# mean
df_exams['math score'].mean().round(10)
```

```python
# standard deviation
df_exams['math score'].std()
```

```python
# max
df_exams['math score'].max()
```

```python
# min
df_exams['math score'].min()
```

## Math operatons in row

```python
# calculating the sum in a row
sum_column = df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
sum_column
```

```python
# insert the series into the dataframe as new column
df_exams.insert(2, 'sum_column', sum_column, allow_duplicates=False)
df_exams
```

```python
# calculating the average result
avg_result = sum_column / 3
avg_result.round(2)
```

```python
# insert this average data as new column average
df_exams.insert(6, 'average', avg_result)
df_exams.round(3)
```
